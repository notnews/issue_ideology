```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import gc
```

```python
t = 3
```

```python
out_df = None
for i, df in enumerate(pd.read_csv('../data/major_%s_news_predict.csv' % t, iterator=True, chunksize=5000)):
    print i, len(df)
    if out_df is None:
        out_df = df
    else:
        new_df = pd.concat([out_df, df])
        del out_df
        del df
        out_df = new_df
    gc.collect()
```

```python
print len(out_df)
out_df.drop_duplicates('text', inplace=True)
out_df
```

```python
out_df.to_csv('../data/major_%s_news_predict_dedup.csv' % t, index=False)
```

```python
out_df = None
try:
    for i, df in enumerate(pd.read_csv('../data/major_%s_news_predict_dedup.csv' % t, iterator=True, chunksize=5000)):
        vect = CountVectorizer(ngram_range=(3, 3), binary=True) 
        X = vect.fit_transform(df.text.astype(str))
        print i, X.shape
        ng_df = pd.DataFrame(X.sum(axis=0).tolist()[0], index=vect.get_feature_names(), columns=[i])
        del df
        del vect
        del X
        if out_df is None:
            out_df = ng_df
        else:
            result = pd.merge(out_df, ng_df, left_index=True, right_index=True, how='outer')
            del out_df
            del ng_df
            out_df = result
        out_df['n'] = out_df.sum(axis=1)
        out_df.drop(i, axis=1, inplace=True)
        print len(out_df)
        print out_df.memory_usage(deep=True)
        gc.collect()
except Exception as e:
    print e
#print out_df
```

```python
out_df[out_df['n'] > 10]
```

```python
out_df[out_df['n'] > 10].to_csv('../data/major_%s_news_3gram_df10.csv' % t, index_label='ngram')
```
