```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('../data/major_5_news_predict.csv')
df
```

```python
df.drop_duplicates('text', inplace=True)
df
```

```python
vect = CountVectorizer(ngram_range=(2, 3), min_df=100) 
vect.fit(df.text)
```

```python
len(vect.get_feature_names())
```

```python
vect.vocabulary_
```

```python
ng = vect.get_feature_names()
ng_df = pd.DataFrame(ng)
ng_df.columns = ['ngram']
ng_df.to_csv('../data/major_5_news_23gram_df100.csv', index=False)
ng_df
```
